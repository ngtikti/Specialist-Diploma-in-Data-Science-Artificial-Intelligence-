```python
import numpy as np
import matplotlib.pyplot as plt


title = "Resale Flat Prices"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()


data = np.genfromtxt('resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv',
                     skip_header=1,dtype=[('month','U50'),('town','U50'), ('flat_type','U50'),('block','f8'),('street_name','U50'),('storey_range','U50'),('floor_area_sqm','f8'), ('flat_model','U50'),('lease_commence_date','f8'), ('remaining_lease','U50'),('resale_price','f8')], 
                                           delimiter=",",missing_values=['na','-'],filling_values=[0]) #to import data and column from csv
#note to self: have to key in all parameters of column, or won't work

print("There are {} rows and {} columns in this dataset".format(len(data),len(data[0])))
print()



set_floor_area_sqm = set(data['floor_area_sqm'])
set_town = set(data['town'])
set_resale_price = set(data['resale_price'])
#set_median_rent = set(data['median_rent'])

print(str(len(set_floor_area_sqm)) + " unique values in floor_area_sqm column")
print(str(len(set_town)) + " unique values in town column")
print(str(len(set_resale_price)) + " unique values in resale_price column")
#print(str(len(set_median_rent)) + " unique values in median_rent column")

print()
#to find correlation btwn sqm and resale price for bedok and yishun maybe
x_sqm_bedok = data[data['town']=='BEDOK']['floor_area_sqm']
y_resale_price_bedok = data[data['town']=='BEDOK']['resale_price']

#criteria = (data['wardclass']=='B1') | (data['wardclass']=='B2' )
x_sqm_yishun = data[data['town']=='YISHUN']['floor_area_sqm']
y_resale_price_yishun = data[data['town']=='YISHUN']['resale_price']

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x_sqm_bedok , y_resale_price_bedok , s=30, c='red', marker="*", label='Bedok')
ax1.scatter(x_sqm_yishun , y_resale_price_yishun , s=10, c='blue', marker="x", label='Yishun')

plt.legend(loc='upper left');

plt.show()


#to display median_rent for diff towns
#print(data[data['town']=='YISHUN']['resale_price'][1004])
#print(len(data[data['town']=='YISHUN']['resale_price']))

labels = list(set(data['town'])) #labels
towns = np.arange(0,len(labels)) #lavels to towns
labels_prices = data[['town','resale_price']] #

prices = labels_prices['resale_price']
#print(prices)

avg_prices = {}

for i in labels:
    pricesfortown = prices[labels_prices['town']==i]
    avg = np.average(pricesfortown)
    #print("Average for town " + i + " is {:.0f}".format(avg))
    avg_prices[i] = avg


# Note this part is to allow the barcharts to be displayed in descending order
from collections import OrderedDict
from operator import itemgetter  
avg_prices = OrderedDict(sorted(avg_prices.items(), key = itemgetter(1), reverse = True))
#print(avg_prices)

plt.figure(1, figsize=(30,30))
barchart = plt.bar(list(avg_prices.keys()), list(avg_prices.values()), color='#d62728')

for i in range(len(barchart)):
    bar = barchart[i]
    x,y  = bar.get_xy()
    h = bar.get_height()
    plt.text(x,h,"{:.0f}".format(list(avg_prices())[i]),fontsize=30)

plt.title('Average Resale Price since 2017 across different towns',fontsize=40)
plt.ylabel('Average Resale Price',fontsize=40)
plt.yticks(fontsize=20)
plt.xticks(towns, labels, fontsize=40,rotation='vertical')


plt.show()

```

```python
#first data set from https://data.gov.sg/dataset/median-rent-by-town-and-flat-type?resource_id=6b1ec2ff-7c38-4ce9-9bbb-af865b4d78cb
import numpy as np
import matplotlib.pyplot as plt


title = "Median rent by Flat Type"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()


data = np.genfromtxt('median-rent-by-town-and-flat-type.csv',
                     skip_header=1,dtype=[('quarter','U50'), ('town','U50'), ('flat_type','U50'), 
                                          ('median_rent','f8')], delimiter=",",missing_values=['na','-'],filling_values=[0]) #to import data and column from csv


print("There are {} rows and {} columns in this dataset".format(len(data),len(data[0])))
print()



set_quarter = set(data['quarter'])
set_town = set(data['town'])
set_flat_type = set(data['flat_type'])
set_median_rent = set(data['median_rent'])

print(str(len(set_quarter)) + " unique values in quarter column")
print(str(len(set_town)) + " unique values in town column")
print(str(len(set_flat_type)) + " unique values in flat_type column")
print(str(len(set_median_rent)) + " unique values in median_rent column")


#boxplot to find how rent prices vary for each room type for the period
null_rows = np.isnan(data['median_rent'])
nonnull_rent_data = data[null_rows==False]
#print(nonnull_rent_data)

labels = list(set(nonnull_rent_data['flat_type']))
labels.sort()
print(labels)
flat_types = np.arange(0,len(labels))
median_rents_for_flat_types = nonnull_rent_data[['flat_type','median_rent']] #replace levels_values with median_rents_for_flat_types
print(median_rents_for_flat_types)

median_rents = median_rents_for_flat_types['median_rent'] #replace values with median_rents


median_rents_2_RM = median_rents[median_rents_for_flat_types['flat_type'] == '2-RM']
median_rents_3_RM = median_rents[median_rents_for_flat_types['flat_type'] == '3-RM']
median_rents_4_RM = median_rents[median_rents_for_flat_types['flat_type'] == '4-RM']
median_rents_5_RM = median_rents[median_rents_for_flat_types['flat_type'] == '5-RM']
median_rents_EXEC = median_rents[median_rents_for_flat_types['flat_type'] == 'EXEC']


rents_combined =[median_rents_2_RM, median_rents_3_RM, median_rents_4_RM, median_rents_5_RM,
                  median_rents_EXEC]
print(len(rents_combined))


plt.figure(2, figsize=(30,30))
plt.title(title,fontsize=60)
plt.ylabel('Median Rent',fontsize=50)
plt.yticks(fontsize=40)
plt.xticks(fontsize=40,rotation='vertical')
bp_dict = plt.boxplot(rents_combined,labels=labels,patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp_dict['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )
    
## change color and linewidth of the whiskers
for whisker in bp_dict['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp_dict['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp_dict['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp_dict['fliers']:
    flier.set(marker='D', color='#e7298a', alpha=0.5)

print(bp_dict.keys())

for line in bp_dict['medians']:
    # get position data for median line
    x, y = line.get_xydata()[1] # top of median line
    # overlay median value
    plt.text(x, y, '%.1f' % y,
         horizontalalignment='center',fontsize=30) # draw above, centered

fliers = []
for line in bp_dict['fliers']:
    ndarray = line.get_xydata()
    if (len(ndarray)>0):
       max_flier = ndarray[:,1].max()
       max_flier_index = ndarray[:,1].argmax()
       x = ndarray[max_flier_index,0]
       print("Flier: " + str(x) + "," + str(max_flier))
     
       plt.text(x,max_flier,'%.1f' % max_flier,horizontalalignment='center',fontsize=30,color='green') 

plt.show()
```
